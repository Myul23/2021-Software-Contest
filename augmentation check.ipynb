```python
import os
import numpy as np
import random
import matplotlib.pyplot as plt
from PIL import Image

total = "D:/AIC/darknet-temp/result"

row, col = 5, 5
f, ax = plt.subplots(row, col, figsize=(10, 20))

path = os.path.join(total, str(8))
datas = os.listdir(path)
samples = random.sample([idx for idx in range(len(datas) // 2)], row * col)
for i in range(row):
    # INDEX -> DATA
    for j in range(col):
        data = datas[samples[i * col + j] * 2]
        img = Image.open(os.path.join(path, data))
        ax[i, j].imshow(img)
        ax[i, j].axis("off")

plt.subplots_adjust(hspace=0.05, wspace=0.05)
plt.show()
```
